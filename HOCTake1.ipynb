```python
import json
import requests
import pandas
```

```python
url = 'https://csgenome.org/api/benchmarks/top500?page=1&limit=25000&columns=computer'
data = requests.get(url)
#print(data.json())
#dictt = dict()
dictt = data.json()
#print(dictt)
print(dictt["data"][0]["computer"])
counts = dict()
for i in range(len(dictt["data"])):
    val = dictt["data"][i]["computer"]
    if val in counts:
        counts[val] = counts[val] + 1
    else:
        counts[val] = 1
#print(counts.values())
#sortedCounts = dict()
sortedCounts = sorted(((value, key) for (key,value) in counts.items()), reverse=True)
print(sortedCounts)
```

```python
idx = []
for x in range(0, len(sortedCounts)):
    idx.append(str(x+1) + ".")
sortedDF = pandas.DataFrame(sortedCounts, index=idx, columns=['Count', 'Computer'])
sortedDF = sortedDF[['Computer', 'Count']]
sortedDF.head(50)
```
